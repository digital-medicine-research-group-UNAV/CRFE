```python
import os
import pandas as pd
import numpy as np
import itertools
import scipy
import csv
import matplotlib.pyplot as plt
import pickle

from collections import Counter
```

```python
def count_elements(arrays):
    
    # Use Counter to count the occurrences of each element in the arrays
    count = Counter()

    for array in arrays:
        count.update(array)

    return count


def ordenar_diccionario(diccionario):

    claves_ordenadas = sorted(diccionario.keys(), key=lambda x: int(x))
    diccionario_ordenado = {clave: diccionario[clave] for clave in claves_ordenadas}
    
    return diccionario_ordenado


def read_csv(filename):
    data = []

    with open(filename, 'r') as file:

        reader = csv.reader(file)
        for row in reader:
            data.append(row)

    return data
```
